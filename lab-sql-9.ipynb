```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.


query6 = '''SELECT title, length, RANK() OVER length as ranking FROM sakila.film
WHERE (length IS NOT NULL) AND (length = '  ');'''

film = pd.read_sql_query(query6, engine)
film.index
film
```
